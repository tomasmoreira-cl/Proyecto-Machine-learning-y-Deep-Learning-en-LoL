**Creaci贸n de columnas**

- Creamos columnas llamadas wr_champ y wr_global, la primera es el win rate del campe贸n que estoy usando, actualiz谩ndose con cada registro del campe贸n (inicializandose en 50%, es decir, ganar o perder es aleatorio). Mientras que la segunda es el wr global al momento del registro anterior (shift = 1), es decir, toma el win rate hist贸rico (promedio acumulado de win/rate) en cada registro.

```python
# IMPORTANTE: El DataFrame DEBE estar ordenado cronol贸gicamente antes de esto.

# Rolling Global Win Rate (W/R actualizado en cada registro)

# Promedio acumulado w/r de todas tus partidas .
# Capturamos rendimiento general, considerando hasta la partida anterior al registro analizado.
df_base['history_wr_global'] = (
    df_base['win']
    .shift(1)             # Ocultar partida actual (Evitar Leakage)
    .expanding()          # Ventana expansiva (toda la historia previa, acumulativo)
    .mean()               # Calcular promedio
    .fillna(0.5)          # Inicializaci贸n neutra para el registro #1
)


# Rolling Champion Win Rate (W/R con cada campe贸n)

# L贸gica: Promedio de win rate acumulado filtrado por campe贸n.
# Objetivo: Saber qu茅 tan bueno eres espec铆ficamente con ese pick.

df_base['history_wr_champ'] = (
    df_base.groupby('champion')['win']                 # Obtener serie de win de cada campe贸n 
    .apply(lambda x: x.shift(1).expanding().mean())    # Promedia las wins hist贸ricas
    .reset_index(level=0, drop=True)                   # Alinear 铆ndice con el DF original
    .fillna(0.5)                                       # Si es la primera vez que juegas un champ, se rellena con 0.5 (Neutro)
)

# VERIFICACIN
# Revisamos un caso de uso real para asegurar que los n煤meros tienen sentido.
cols_debug = ['match_num', 'champion', 'win', 'history_wr_global', 'history_wr_champ']
champ_debug = df_base['champion'].mode()[0] # Tu main

print(f"--- Revisi贸n de Datos para {champ_debug} ---")
print(df_base[df_base['champion'] == champ_debug][cols_debug].head(5))

print("\n--- Revisi贸n de Inicio de Datos (Primeras partidas absolutas) ---")
print(df_base[cols_debug].head(5))
```

**Creaci贸n de columnas**

Creamos las columnas CS_per_min, KDA, total_wards_placed,vs_per_min para un mejor an谩lisis descriptivo de las partidas. Luego las eliminaremos para el modelamiento para evitar ruido.

```python
# L贸gica = Subditos asesinados + monstruos neutrales asesinados / duraci贸n = creeps por minuto
# Obejtivo = Usar esta m茅trica en el an谩lisis descriptivo
df_base['CS_per_min'] = (
    (df_base['lane_minions'] + df_base['jungle_minions']) / 
     df_base['game_duration'].replace(0, 1)
    ).astype(float).round(2)

# L贸gica = Asesinatos +  asistencias / muertes = KDA. (reemplazamos muertes 0 por 1 para evitar errores)
# Obejtivo = Usar esta m茅trica en el an谩lisis descriptivo
df_base['KDA'] = (
    (df_base['kills'] + df_base['assists']) / 
    df_base['deaths'].replace(0, 1)
    ).astype(float).round(2)

# L贸gica = Centinelas colocados +  Centinelas de control colocados. (total centinelas)
# Obejtivo = Usar esta m茅trica en el an谩lisis descriptivo
df_base['total_wards_placed'] = df_base['wards_placed'] + df_base['control_wards_placed']
df_base['wards_per_min'] = df_base['total_wards_placed'] / df_base['game_duration'].round(2)

# L贸gica = Centinelas colocados +  Centinelas de control colocados. (total centinelas)
# Obejtivo = Usar esta m茅trica en el an谩lisis descriptivo
df_base['vs_per_min'] = df_base['vision_score'] / df_base['game_duration'].round(2)

# L贸gica: Creamos una copia del data frame y filtros por tipo de cola para el an谩lisis de datos.
# Objetivo: Tener dataframes distintos para el modelamiento y para el an谩lisis de datos.
df_analytics = df_base.copy()
df_analytics_soloq = df_base[df_base['queue_id']== 420] 
df_analytics_flex = df_base[df_base['queue_id']== 440] 


# Convertimos la columna categ贸rica 'role' en variables binarias (dummies)
df_base = pd.get_dummies(
    df_base, 
    columns=['role'], 
    drop_first=True, 
    dtype=int)

df_base['is_flex'] = (df_base['queue_id'] == 440).astype(int)
df_base = df_base.drop(columns = ['queue_id'])

# VERIFICACIN
# Busca las nuevas columnas creadas
new_cols = [col for col in df_base.columns if 'role_' in col or 'CS' in col or 'KDA' in col or 'is_' in col] 
print(f"Nuevas columnas de rol creadas: {new_cols}")
df_base[new_cols].head()
```

## An谩lisis comportamiento del sistema: Elasticidad y "Loss Q" (correcci贸n de MMR)

M谩s all谩 del rendimiento individual por campe贸n, es cr铆tico analizar el comportamiento del algoritmo de emparejamiento (Matchmaking).

El Objetivo: Investigar estad铆sticamente la existencia de patrones de correcci贸n forzada (coloquialmente conocidos como "Losers Queue"). Se busca determinar si el sistema impone una resistencia artificial tras rachas de victorias para forzar una regresi贸n al 50% de Win Rate.

**驴El estancamiento en la liga se debe a un l铆mite de habilidad personal o a un techo artificial algor铆tmico que penaliza las rachas positivas excesivas?**

### Elasticidad Partidas Flex

```python
# Creamos un agrupador de rachas y asignamos un identificador (comienza en 1). 
# Si el resultado actual es distinto al anterior, empieza nueva racha, por lo tanto, un nuevo identificador.
df_analytics_flex['streak_group'] = (df_analytics_flex['win'] != df_analytics_flex['win'].shift(1)).cumsum()
df_analytics_flex['streak_counter'] = df_analytics_flex.groupby('streak_group').cumcount() + 1

# Obtenemos la racha actual y determinamos si es racha de derrotas o de victorias
# Multiplica por 1 o -1. Cuando se gana se deja como racha de victorias (1,2,3...), cuando no, como derrotas (-1,-2,-3...). 
df_analytics_flex['racha_actual'] = np.where(
    df_analytics_flex['win'] == 1, 
    df_analytics_flex['streak_counter'],      # Caso Win: 3
    df_analytics_flex['streak_counter'] * -1  # Caso Loss: -3
)

# 2. Predecir el FUTURO (Target Leakage intencional para an谩lisis)
# Queremos saber: "Dada mi racha actual, 驴gan茅 la SIGUIENTE?"
df_analytics_flex['next_game_win'] = df_analytics_flex['win'].shift(-1)

# Para graficar
# Agrupamos por la racha actual y obtenemos el promedio de ganar la siguiente partida y la frecuencia de las racha. 
streak_stats_flex = df_analytics_flex.groupby('racha_actual')['next_game_win'].agg(['mean', 'count'])
streak_stats_flex = streak_stats_flex[streak_stats_flex['count'] >= 5].round(2) # M铆nimo 5 casos para que sea estad铆sticamente v谩lido

print("--- Probabilidad de Ganar la SIGUIENTE partida seg煤n tu Racha Actual ---")
print(streak_stats_flex)


# Graficamos los datos obtenidos
plt.figure(figsize=(10, 6))
plt.plot(streak_stats_flex.index, streak_stats_flex['mean'], marker='o', linestyle='-', color='purple')
plt.axhline(0.5, color='gray', linestyle='--', label='Azar (50%)')
plt.title('驴El sistema te frena? (Probabilidad de Victoria vs Racha Actual Flex)')
plt.xlabel('Racha Actual (Negativo = Derrotas seguidas, Positivo = Victorias seguidas)')
plt.ylabel('Probabilidad de ganar la SIGUIENTE partida')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```
